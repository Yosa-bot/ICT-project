```python
def tax_none(income):
    print(income * 0)

def first_tax(income):
    print(income * 0.15)

def second_tax(income):
    print(income * 0.2)

def third_tax(income):
    print(income * 0.25)

def fourth_tax(income):
    print(income * 0.3)

def fifth_tax(income):
    print(income * 0.35)

income = float(input("Please enter your income to calculate yoiur tax: "))

if income <= 2000:
    tax_none(income)
elif income <= 4000:
    first_tax(income)
elif income <= 7000:
    second_tax(income)
elif income <= 10000:
    third_tax(income)
elif income <= 14000:
    fourth_tax(income)
elif income > 14000:
    fifth_tax(income)
else: 
    print("Invlid in")
```
